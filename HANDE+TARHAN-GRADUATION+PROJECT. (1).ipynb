# 2.TIME SERIES 
**Definition** : The series are called the "time series" when observations are made at certain times(time-sequenced) such as day, month, year, minute, second, and millisecond. It is a set of time-sequenced measurements of a size that is of interest. The purpose of this analysis of the time series is to understand the reality represented in the observation set. A time dependent variable is represented by time series charts sorted by time

Before the econometric methods are applied to the time series, it is necessary to examine the different elements that constitute the time series. 

## 2.1. Components of Time Series

A time series decomposition is a mathematical operation which convert a time series into multiple different time series. The original time series is usually decomposed into 3 sub-time series:

#### 2.1.1. Trend: 
The underlying trend of the measurements.It is a steady state that occurs after long periods of fall and elevation of time series. Time series has a tendency to decay or rise steadily from long term perspective. It does not have to be linear and Sometimes it can go from an increasing trend to a downward trend.

#### 2.1.2. Seasonal: 
Increase or decrease periodically in a special time period within a time series such as 3 months, monthly or weekly brings seasonal changes in time series.A seasonal pattern exists when a series is influenced by seasonal factors. Seasonality is always of a fixed and known period.

#### 2.1.3. Cycle
Periodic changes that are not related to seasonal changes. For example, the cyclic process describes short-term expansion or contraction in the economy, independent of the general trend.These fluctuations usually long at least 2 years.

#### 2.1.4. Random: ( or “Remainder”)
Is the residuals of the time series after splitting into the seasonal and trends time series. It is also called "the error".

## 2.2.Additive and Multiplicative Time Series

It is significant to choose between the additive or multiplicative model to get a useful decomposition. 

### 2.2.1. Additive Time Series
If a time series model is 'additive' , if data model in which the effects of individual factors are differentiated and added to model the data. the magnitude of seasonal variation must be constant.
#### $$Formula: Time Series = Seasonal + Trend + Random$$

### 2.2.2. Multiplicative Time Series
If the time series model is 'multiplicative', if the trend and seasonal components are multiplied and then added to the error component.The magnitude of seasonal variation must increase when the time series increases.
#### $$ Formula: Time Series = Seasonal \times Trend \times Random $$

An alternative way of using a multiplicative model is to transform the multiplicative data to additive data with using **log**.  and then use an additive model.
$Time Series = Seasonal \times Trend \times Random$, equivalent to $log(TimeSeries) = log(Seasonal) + log(Trend)+ log(Random)$.

##### When to apply an additive decomposition?
1-The behaviors of the components are independent from each other. For example, an increase in the trend-cycle will not lead to an increase in the magnitude of seasonal downs.

2-The difference of the trend and the main data is almost constant in similar periods of time regardless of the tendency of the trend.

3-The seasonal movements are the approximately stable.

##### When to apply a multiplicative decomposition?
The multiplicative model is suitable if the tendency of the trend leads to changes in the seasonal and irregular fluctuations in a specific manner. Hence, unlike the additive case, the components are not independent from each other.

## 3. ARIMA (Auto-Regressive Integrated Moving Average )

Before we introduce ARIMA models, we need to first discuss the concept of stationarity.

### 3.1. Stationarity:

A series is said to be stationary when its mean, variance, and covariance are not time-dependent. The modeling of stable series with consistent features has less uncertainty since ARIMA uses the delay of the previous series to model its behavior.If the time series is not stationary it will include trend.

$$ E(y_t) = E(y_{t−s}) = 𝜇  $$

$$E[(y_t − 𝜇)^2] = E[(y_{t−s} − 𝜇)^2] = \sigma^2_y $$  $  [var(y_t) = var(y_{t−s}) = \sigma^2_y ]$

$$ E[(y_t − 𝜇 )(y_{t−s} − 𝜇)] = E[(y_{t−j} − 𝜇)(y_{t−j−s} − 𝜇)] = \gamma_s $$
$ [cov(y_t, y_{t−s}) = cov(y_t−j, y_{t−j−s}) = \gamma_s] $

where 𝜇 , $ \sigma^2, \gamma_s $ are constants.


ARIMA is defined as 3 parameters: ARIMA **_(p,d,q)_**. It has three stages: AR(AutoRegressive), I(Integrated) and MA(Moving Average). Purpose of each three step is to provide residuals display a *white noise* pattern or no pattern at all.

### 3.2. I(Integrated)
The *d* specifies the degree of differencing in the integrated **I(d)** component. Differencing a series involves subtracting its current and previous values d times until make it stationary. Usually time series become trendless after 1st difference.

$$ Y_t'= Y_t - Y_{t-1} $$

###  3.3. AR(AutoRegressive )
An auto regressive **AR(p)** component is representing to the use of past values in the regression equation for the series Y. The auto-regressive parameter *p* identifies the number of lags used in the model. For example, AR(2) or, equivalently, ARIMA(2,0,0), is represented as

$$Y_t = c + \phi_1y_{t-1} + \phi_2 y_{t-2}+ e_t$$

where $ \phi_1, \phi_2 $ are parameters for the model.

### 3.4. MA(Moving Average)

A moving average **MA(q)** component represents the error of the model which is a combination of previous error terms $e_t$. The order *q* identifies the number of terms to include in the model

$$Y_t = c + \theta_1 e_{t-1} + \theta_2 e_{t-2} +...+ \theta_q e_{t-q}+ e_t$$

A non-seasonal $ARIMA$ model is constructed by differencing, autoregressive, and moving average components. It can be written as a linear equation:

$$ Y_t = c + \phi_1y_d{_{t-1}} + \phi_p y_d{_{t-p}}+...+\theta_1 e_{t-1} +  \theta_q e_{t-q} + e_t$$

 $y_d$ : Y differenced d times
 
 c : constant.


## 4. Autocorrelation
*Correlation* evaluate the linear relation between a pair of variables and defined by

$$ cor(x, y) =  \frac { E((x − μ_x)(y − μ_y))}{ \sigma_x \sigma_y }= \frac{cov(x, y)}{\sigma_x \sigma_y} $$

Correlation takes a value between −1 and +1. A value of 0 demonstrates there is no linear association. On the other hand a value of +1 means there is an linear association between the pairs and it lying on a straight line of positive slope. Likewise, -1 means there is an linear association between the pairs with lying on a straight line of negative slope.

**Autocorrelation** formula occurs  by replacing $x$ with $x_t$ in the formula of correlation and $y$ with $x_{t+k}$, and using the overall mean $μ_x$ for the mean of both $x_t$ and $x_{t+k}$. 

$$ \gamma_k = cor(x_t, x_{t+k}) = \frac{E( (x_t − μ_x)(x_{t+k} − μ_x) )}{2x} $$
Note that $\gamma_0$ is always 1.

## 4.1. Correlograms

### 4.1.1. ACF (Autocorrelation Function) Plots
ACF plots demonstrates correlation between a series and its lags. Besides it is beneficial for determining whether a series is stationary, ACF plots are useful when suggesting the order of differencing and determining the order of the *MA (q)* model.

The x-axis identifies the lag (k) and the y-axis identifies the autocorrelation ($ \gamma_k$) at each lag. This means the correlogram is a plot of $\gamma_k$ against k. The dotted lines represent the % significance level for the statistical test $ H_0 : \gamma_k$ = 0. If a correlation that goes beyond these lines, it is "significantly" different from zero. The lag 0 autocorrelation is always 1 and is demonstated on the plot.

At the bottom we can see the line which significantly goes beyond the dotted line is "1". Then our series form a model of MA(1).

### 4.1.2. PACF(Partial Autocorrelation Function) Plots
Partial autocorrelation plots (PACF),  shows correlation between a variable and its lags that is not clarified by previous lags. They are helpful when determining the order of the AR(p) model. In summary, in time series analysis, the partial autocorrelation at lag k  is defined as the autocorrelation between $X_t$ and $X_{t-k} $ that is not accounted for by lags t+1 through t+k-1. It means that in comparison with autocorrelation, in partial autocorrelation the correlations with all the elements within the lag are removed. It is important that following the definition, partial autocorrelation for lag 1 is equivalent to autocorrelation for lag 1.

### 4.1.3. Reading the Plots

The dotted lines in ACF and PACF graphs are %significance limit. The area under the dotted line is called 'insignificant area'. If only the first lag of the PACF graph is statistically significant, that means only the first lag exceeds the confidence limit, and the relationship quantities in the ACF graph decrease gradually as the number of lags increases, the most suitable model is the AR(1) model. If the first 'p' lines exceed the confidence limit, then this model is AR(p). Likewise, if only the first 'q' lag of the ACF graph is statistically significant, that is, only the 'q' number of lines exceed the confidence limit, and the relationship quantities in the PACF graph decrease gradually as the number of lags increases, the most suitable model MA(q).

```r
acf(ts(random_additive))
pacf(ts(random_additive))
```

 ## 5. auto.arima( ) function
 R provides  *forecast* package to the user for clearly identfy the order of the model using the $arima()$ function, or automatically generate a set of optimal (p, d, q) using **auto.arima()**. This function searches through combinations of order parameters and selects the set that optimizes the model fit metrics.

The auto.arima() function in R uses a variation of the Hyndman and Khandakar algorithm which combines unit root tests, minimization of the AICc and MLE to obtain an ARIMA model. 


There exist a large number of  criterions for comparing quality of models with each other. Two of the most commonly used are Akaike information criteria (AIC) and Baysian information criteria (BIC). These criteria are associated and they can estimate amount of lost information because of the selected model.When comparing models, one wants to minimize AIC and BIC.

They are defined as follows:

### 5.1. The Corrected Akaike Information Criterion(AICc) and Baysian information criteria (BIC)

The Akaike information criterion (AIC) is a quality statistical relative model measure for a given set of data. That is, when a collection of data models is given, the AIC estimates each model quality, each of the other models relatively. Thus, the AIC provides a way to select the model. The Akaike Criterion does not provide a model test for null hypothesis testing; that means the it can say something about the quality of the model in an absolute sense. If all candidate models are bad, it will not give any warning. 

$$ AICc= -2LogLikelihood + 2k + 2k(k+1) / (n-k-1) $$
where k is the number of estimated parameters in the model and n is the number of observations in the data set. 

$$BIC= -2LogLikelihood + k(ln(n)) $$
When comparing the AICc and BIC values for two models, the model with the smaller value is considered better.

### 5.2.Maximum Likelihood Estimation (MLE)

When R predicts the ARIMA model, it uses the maximum likelihood Estimation(MLE). This technique finds the values of the parameters that maximize the likelihood of obtaining the observed data. For ARIMA models, MLE is very similar to estimates of the smallest squares by minimizing
$$ \sum_{t=1}^{T} \left(e_t^2 \right) $$


```r
library(forecast)
auto.arima(random_additive, seasonal=FALSE)
```

$$ \hat Y_{d_t} = 0.6634 Y_{t-1} -0.3897 Y_{t-2}+ 0.4595 Y_{t-3} -0.5610Y_{t-4}+ 0.0911 e_{t-1} + 0.3046 e_{t-2} -0.6896 e_{t-3}  + E$$
 
###### where E is some error and the original series is differenced with order 0.

## 6. Kolmogorov - Smirnov Test

Kolmogorov-Smirnov Test is a statistical,non-parametric test for normality which compares the actual data with the normal distribution. It is applicable only for continuous functions and it tends to be more sensitive near the center of distribution than at the tails.Normality (normal distribution suitability) is very important in practice of statistical confidence intervals.

### 6.1. One-Sample K-S Test

It tests whether an observed frequency distribution matches normal distribution. That is, it points out whether it is reasonable to consider that the scores observed in the sample may have come from a society whose theoretical distribution is obvious.

### 6.2.Two-Sample  K-S Test
The two sample test is a test that examines whether two independent samples come from the same population (or from the same population). The two-sided test is sensitive to all kinds of variances (skewness, kurtosis, curvature) in the distributions of the samples. 

The hypotheses used are:

*$H_0(null hypothesis)$*       : The sample data are normally distributed.

*$H_a(alternative hypothesis)$*: The sample data are not normally distributed.

*$D$*: maximum absolute difference between the normal population and the given data.


$D=  max_{N \ge y \ge 1}   [F(Y_i)  - \frac{i-1}{n}, \frac{i}{n} - F(Y_i) ] $


*$p-value$*: significance value of the test.

• The test rejects the null-hypothesis when the p-value is less than 0.05.

• The test fail the reject the null hypothesis(i.e.normal distribution may be assumed) when the p-value is greater than 0.05.

```r
ks.test(x_add,rnorm(500))
```

That means the distribution is very different than normal distribution and the test is very sure about that. Therefore it rejects the null hypothesis which is $H_0$ = normal distribution

## 7. STL: A Seasonal-Trend Decomposition Procedure Based on LOESS

STL, "Seasonal and Trend Decomposition using Loess" is a filtering procedure used to estimate nonlinear relationships and  decomposes a time series into seasonal, trend and irregular components using loess.Having not only certain frequency, being able to work for changing seasonality and controlling the rate of change and smoothness by user are the advantages of this method. On the other hand is works more complicated and slower.STL decomposition works with additive time series by default so, we log will br taken fot transform the multiplicative series to additive series.


### 7.1. Loess

Loess ("locally-weighted scatterplot smoothing") uses local regression to remove "jaggedness" from data. A window of a certain width is placed over the data. The wider the window, the smoother the resulting loess curve becomes. A regression line (or curve) is placed in the observations falling into the window; the points closest to the center of the window are weighted to have the greatest effect on the calculation of the regression line. The weighting is reduced on the farthest points from the window's regression line. The regression is restarted and the weights are recalculated. This process is repeated several times. So, we get a point on the loess curve. This is the point on the regression line in the middle of the window. Loess is calculated by moving the curve over the window's data. Each point on the resulting root curve is the intersection of a regression line and a vertical line in the middle of such a window.
