```r
x=Quandl("CURRFX/EURRUB", collapse="monthly", start_date="2000-01-01", end_date="2016-12-31",type="ts")
x
```

```r
plot(x)
```

```r
class(x)
```

```r
decompose(x)
```

```r
plot(decompose(x))
```

```r
y=decompose(x)
y$trend
```

```r
plot(y$trend)
```

```r
trend= filter(x,filter =c(1/24,1/12,1/12,1/12,1/12,1/12,1/12,1/12,1/12,1/12,1/12,1/12,1/24),sides=2)
trend=ts(trend, freq=12)
plot(trend, xlab="trend")
```
